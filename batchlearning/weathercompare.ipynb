```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the datasets
df1 = pd.read_csv('forecast.csv')
df2 = pd.read_csv('weatherhistoricalweek.csv')
```

```python
df1['time'] = pd.to_datetime(df1['time'])
df2['time'] = pd.to_datetime(df2['time'])
```

```python
# Merge the datasets on the 'date' column
merged = pd.merge(df1, df2, on='time', suffixes=('_df1', '_df2'))
```

```python
# Plot the temperature from both datasets
plt.figure(figsize=(10, 6))
plt.plot(merged['time'], merged['temperature_2m (°C)_df1'], label='forecast')
plt.plot(merged['time'], merged['temperature_2m (°C)_df2'], label='historical')
plt.xlabel('time')
plt.ylabel('Temperature')
plt.title('Temperature Comparison')
plt.legend()
plt.show()
```

```python
# Plot the cloud cover from both datasets
plt.figure(figsize=(10, 6))
plt.plot(merged['time'], merged['cloud_cover (%)_df1'], label='Cloud Cover - Dataset 1')
plt.plot(merged['time'], merged['cloud_cover (%)_df2'], label='Cloud Cover - Dataset 2')
plt.xlabel('Date')
plt.ylabel('Cloud Cover')
plt.title('Cloud Cover Comparison')
plt.legend()
plt.show()
```

```python
# Plot the sunshine from both datasets
plt.figure(figsize=(10, 6))
plt.plot(merged['time'], merged['sunshine_duration (s)_df1'], label='Sunshine - Dataset 1')
plt.plot(merged['time'], merged['sunshine_duration (s)_df2'], label='Sunshine - Dataset 2')
plt.xlabel('Date')
plt.ylabel('Sunshine')
plt.title('Sunshine Comparison')
plt.legend()
plt.show()
```

```python
# Calculate the difference between the two datasets
merged['temperature_2m_diff'] = merged['temperature_2m (°C)_df1'] - merged['temperature_2m (°C)_df2']
merged['cloud_cover_diff'] = merged['cloud_cover (%)_df1'] - merged['cloud_cover (%)_df2']
merged['sunshine_diff'] = merged['sunshine_duration (s)_df1'] - merged['sunshine_duration (s)_df2']
```

```python
# Plot the temperature difference
plt.figure(figsize=(10, 6))
plt.plot(merged['time'], merged['temperature_2m_diff'], label='Temperature Difference')
plt.xlabel('Date')
plt.ylabel('Temperature Difference')
plt.title('Temperature Difference Comparison')
plt.legend()
plt.show()

# Plot the cloud cover difference
plt.figure(figsize=(10, 6))
plt.plot(merged['time'], merged['cloud_cover_diff'], label='Cloud Cover Difference')
plt.xlabel('Date')
plt.ylabel('Cloud Cover Difference')
plt.title('Cloud Cover Difference Comparison')
plt.legend()
plt.show()

# Plot the sunshine difference
plt.figure(figsize=(10, 6))
plt.plot(merged['time'], merged['sunshine_diff'], label='Sunshine Difference')
plt.xlabel('Date')
plt.ylabel('Sunshine Difference')
plt.title('Sunshine Difference Comparison')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Plot the temperature from both datasets on the first subplot
axs[0].plot(merged['time'], merged['temperature_2m (°C)_df1'], label='forecast')
axs[0].plot(merged['time'], merged['temperature_2m (°C)_df2'], label='historical')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Temperature')
axs[0].set_title('Temperature Comparison')
axs[0].legend()

# Plot the temperature difference on the second subplot
axs[1].plot(merged['time'], merged['temperature_2m_diff'], label='Temperature Difference', color='red')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Temperature Difference')
axs[1].set_title('Temperature Difference Comparison')
axs[1].legend()

# Adjust the layout
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Plot the cloud cover from both datasets on the first subplot
axs[0].plot(merged['time'], merged['cloud_cover (%)_df1'], label='forecast')
axs[0].plot(merged['time'], merged['cloud_cover (%)_df2'], label='historical')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Cloud Cover')
axs[0].set_title('Cloud Cover Comparison')
axs[0].legend()

# Plot the cloud cover difference on the second subplot
axs[1].plot(merged['time'], merged['cloud_cover_diff'], label='Cloud Cover Difference', color='red')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Cloud Cover Difference')
axs[1].set_title('Cloud Cover Difference Comparison')
axs[1].legend()

# Adjust the layout
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Plot the sunshine from both datasets on the first subplot
axs[0].plot(merged['time'], merged['sunshine_duration (s)_df1'], label='forecast')
axs[0].plot(merged['time'], merged['sunshine_duration (s)_df2'], label='historical')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Sunshine')
axs[0].set_title('Sunshine Comparison')
axs[0].legend()

# Plot the sunshine difference on the second subplot
axs[1].plot(merged['time'], merged['sunshine_diff'], label='Sunshine Difference', color='red')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Sunshine Difference')
axs[1].set_title('Sunshine Difference Comparison')
axs[1].legend()

# Adjust the layout
plt.tight_layout()
plt.show()
```
