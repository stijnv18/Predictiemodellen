```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("Solar_data.csv")

# Drop all colums except BE


df_be = df['BE']

# Plot the 'BE' column
plt.figure(figsize=(14, 7))
df_be.plot()
plt.title('Energy Production of Solar - BE')
plt.xlabel('Time')
plt.ylabel('Energy Production')
plt.show()


df_be_week = df_be.tail(168)

# Plot the 'BE' column in gigawatt hours
plt.figure(figsize=(14, 7))
df_be_week.plot()
plt.title('Energy Production of Solar - BE (One Week)')
plt.xlabel('Time (Hourly Intervals)')
plt.ylabel('Energy Production')
plt.show()

```
