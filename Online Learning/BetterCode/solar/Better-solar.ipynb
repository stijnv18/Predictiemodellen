```python
import pandas as pd

# Load the data
df = pd.read_csv('Plant_1_Generation_Data.csv')

# Convert DATE_TIME to datetime
df['DATE_TIME'] = pd.to_datetime(df['DATE_TIME'], format='%d-%m-%Y %H:%M')

# Aggregate data by DATE_TIME
df_agg = df.groupby('DATE_TIME').agg({'DC_POWER': 'sum', 'AC_POWER': 'sum'})

print(df_agg)
    
```
